```python
class CafeManagementSystem:                                                             # Define a class
    def __init__(self):                                                                 # Constructor -> super power 
        self.menu = {                                                                   # Self is an object 
            "Coffee": 3.0,
            "Tea": 2.5,
            "Sandwich": 5.0,
            "Cake": 4.0,
            "Juice": 3.5
        }
        self.order = {}                                                                 # passing a dict 
    
    def display_menu(self):                                                             # Define a method 
        print("\n--- Cafe Menu ---")
        for item, price in self.menu.items():                                           # Use for loops 
            print(f"{item}: ${price:.2f}")
    
    def take_order(self):                                                               # Define a method 
        while True:                                                                     # Use a loops 
            self.display_menu()
            choice = input("Enter item to order (or type 'done' to finish): ").title()
            
            if choice == 'Done':                                                        # use a conditional statements 
                break
            
            if choice in self.menu:
                quantity = int(input(f"Enter quantity for {choice}: "))
                if choice in self.order:
                    self.order[choice] += quantity
                else:
                    self.order[choice] = quantity
                print(f"{quantity} {choice}(s) added to your order.\n")
            else:
                print("Invalid item. Please choose from the menu.\n")
    
    def generate_bill(self):                                                        # define a metrhod 
        print("\n--- Order Summary ---")
        total = 0                                                                   # use loops 
        for item, quantity in self.order.items():
            price = self.menu[item] * quantity
            print(f"{item} x{quantity} - ${price:.2f}")
            total += price
        print(f"Total Bill: ${total:.2f}")

    def run(self):                                                                  # define a method 
        print("Welcome to the Cafe!")                                               # use print that 
        self.take_order()
        if self.order:                                                              # use if _> codition is true -> executed 
            self.generate_bill()
            print("\nThank you for visiting! Have a great day!")
        else:
            print("No items ordered. Goodbye!")

if __name__ == "__main__":                                                          # use main function -> to execute the whole programm 
    cafe = CafeManagementSystem()
    cafe.run()
```
