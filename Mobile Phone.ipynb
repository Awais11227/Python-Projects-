```python
import time
import random

class MobilePhone:                                                  # defining a class 
    def __init__(self, brand, model, battery=100):                  # constructor - > super power
        self.brand = brand                                          # self is an object
        self.model = model
        self.battery = battery
        self.contacts = {}
        self.messages = []
        self.apps = ["Camera", "Gallery", "Calculator", "WhatsApp", "Instagram"]
        self.security_features = ["Face ID", "Fingerprint Scanner"]
        self.internet_speed = "5G"
        self.wallet = True

    def show_battery(self):                                          # define a method
        print(f"Battery Level: {self.battery}%")

    def charge_phone(self):                                          # define a method
        self.battery = 100
        print("Phone is fully charged!")

    def make_call(self, number):                                     # define a method
        if self.battery < 10:
            print("Battery too low to make a call!")
        else:
            print(f"Calling {number}...")
            self.battery -= 10

    def send_message(self, number, text):                            # define a method
        if self.battery < 5:
            print("Battery too low to send a message!")
        else:
            self.messages.append((number, text))
            print(f"Message sent to {number}: {text}")
            self.battery -= 5

    def add_contact(self, name, number):                            # define a method and passing parameters 
        self.contacts[name] = number
        print(f"Contact {name} added successfully!")

    def open_app(self, app_name):                                   # define a method
        if self.battery < 15:
            print("Battery too low to open apps!")
        elif app_name in self.apps:
            print(f"Opening {app_name}...")
            time.sleep(2)
            self.battery -= 15
        else:
            print("App not found!")

    def show_contacts(self):                                     # define a method
        if not self.contacts:
            print("No contacts saved!")
        else:
            print("Contacts:")
            for name, number in self.contacts.items():
                print(f"{name}: {number}")

    def check_messages(self):                                    # define a method
        if not self.messages:
            print("No messages available!")
        else:
            print("Messages:")
            for num, msg in self.messages:
                print(f"From {num}: {msg}")

    def show_security_features(self):                            # define a method
        print("Security Features:", ", ".join(self.security_features))

    def show_internet_speed(self):                               # define a method
        print(f"Internet Speed: {self.internet_speed}")

    def show_wallet_status(self):                                # define a method
        print("Mobile Wallet Enabled" if self.wallet else "Mobile Wallet Disabled")

# User Interaction
brand = input("Enter mobile brand: ")                              # use user input
model = input("Enter mobile model: ")
phone = MobilePhone(brand, model)

while True:                                                     # use loops 
    print("\n--- Mobile Phone Menu ---")
    print("1. Show Battery Level")
    print("2. Charge Phone")
    print("3. Make a Call")
    print("4. Send a Message")
    print("5. Add a Contact")
    print("6. Open an App")
    print("7. Show Contacts")
    print("8. Check Messages")
    print("9. Show Security Features")
    print("10. Show Internet Speed")
    print("11. Show Wallet Status")
    print("12. Exit")
    
    choice = input("Enter your choice: ")                           # enter your choice 
    
    if choice == "1":
        phone.show_battery()
    elif choice == "2":
        phone.charge_phone()
    elif choice == "3":
        num = input("Enter number to call: ")
        phone.make_call(num)
    elif choice == "4":
        num = input("Enter number to send message: ")
        msg = input("Enter your message: ")
        phone.send_message(num, msg)
    elif choice == "5":
        name = input("Enter contact name: ")
        num = input("Enter contact number: ")
        phone.add_contact(name, num)
    elif choice == "6":
        app = input("Enter app name to open: ")
        phone.open_app(app)
    elif choice == "7":
        phone.show_contacts()
    elif choice == "8":
        phone.check_messages()
    elif choice == "9":
        phone.show_security_features()
    elif choice == "10":
        phone.show_internet_speed()
    elif choice == "11":
        phone.show_wallet_status()
    elif choice == "12":
        print("Exiting... Goodbye!")
        break
    else:
        print("Invalid choice! Try again.")
```
